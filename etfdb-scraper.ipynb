```python
def scrapeEtfDb(df, etf):
    page = requests.get(baseUrl + etf + "/#esg", headers=headers)
    if page.status_code == 200: 

        soup = BeautifulSoup(page.text, 'html.parser')
        
        data = soup.find('h1', {'class':'data-title'}).text.strip('\n')
        data = data.split('\n',1)[1]

        if "Title" not in df:
            df["Title"] = ""
        df["Title"][etf] = data

        # Add performance to data frame
        elements = soup.find_all(class_='col-xs-6 col-md-3 col-lg-3')
        for element in elements:
            name = element.find(class_="row relative-metric-header-thumb").text.strip('\n')
            data = element.find(class_="relative-metric-bubble-data")
            if data is None:
                data = element.find(class_="relative-metric-bubble-name")
            if name not in df:
                df[name] = ""
            df[name][etf] = data.text.strip('\n')

        # Add dividend info to data frame
        dividend = soup.find(id='dividend')
        elements = dividend.find_all('div', {'class':'col-md-4'})
        for element in elements:
            divs = element.find_all('div', {'class':'text-center'})
            name = divs[0].text.strip('\n')
            data = divs[1].text.strip('\n')
            if name not in df:
                df[name] = ""
            df[name][etf] = data

        # yield 
        element = dividend.find('div', {'class':'row relative-metric-m-top'})
        name = element.find('div', {'class':'relative-metric-header col-xs-4'}).text.strip('\n')
        data = element.find('span', {'class':'relative-metric-bubble-data'})
        if data is None:
                data = element.find('span', {'class':'relative-metric-bubble-name'})
        if name not in df:
            df[name] = ""
        df[name][etf] = data.text.strip('\n')

        # rating
        elements = soup.findAll('tr', id=lambda x: x and x.startswith('rc-1'))
        for element in elements:
            name = element.find('td', {'data-th':'Metric'}).text.strip('\n')
            data = element.find('td', {'data-th':'Metric Realtime Rating'}).text.strip('\n')
            if name not in df:
                df[name] = ""
            df[name][etf] = data

        # Add esg scores to data frame
        elements = soup.find_all('div',{'class':'col-md-4 col-xs-4 score-block'})
        for element in elements:
            if element:
                name = element.find(class_="score-name").text.strip('\n')
                data = element.find(class_="score").text.split('/',1)[0]
                if name not in df:
                    df[name] = ""
                df[name][etf] = data

        # Add ESG and all kind of list elements to data frame
        elements = soup.find_all('li')
        for element in elements:
            pair = element.find_all('span')[0:2]
            if len(pair) >= 2:
                name = pair[0].text.strip('\n')
                data = pair[1].text.strip('\n')
                if name not in df:
                    df[name] = ""
                df[name][etf] = data

        # Add holdings to data frame
        elements = soup.find_all("table", {"class":'chart base-table'})
        if elements:
            assetList = json.loads(elements[0]["data-chart-series"])
            for element in assetList:
                name = 'asset_' + element['name'].strip('\n')
                data = element["data"][0]
                if name not in df:
                    df[name] = ""
                df[name][etf] = data

            sectorList = json.loads(elements[1]["data-chart-series"])
            for element in sectorList:
                name = 'sector_' + element['name'].strip('\n')
                data = element["data"][0]
                if name not in df:
                    df[name] = ""
                df[name][etf] = data

            capList = json.loads(elements[2]["data-chart-series"])
            for element in capList:
                name =  'cap_' + element['name'].strip('\n')
                data = element["data"][0]
                if name not in df:
                    df[name] = ""
                df[name][etf] = data

            regionList = json.loads(elements[3]["data-chart-series"])
            for element in regionList:
                name = 'region_' + element['name'].strip('\n')
                data = element["data"][0]
                if name not in df:
                    df[name] = ""
                df[name][etf] = data

            marketList = json.loads(elements[4]["data-chart-series"])
            for element in marketList:
                name = 'market_' + element['name'].strip('\n')
                data = element["data"][0]
                if name not in df:
                    df[name] = ""
                df[name][etf] = data
    return df
```
